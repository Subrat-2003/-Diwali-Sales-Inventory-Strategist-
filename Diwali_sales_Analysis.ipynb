```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_diwali_sales = pd.read_csv(r"C:\Users\subra\Downloads\Diwali Sales Data.csv", encoding='cp1252')
df_diwali_sales
```

```python
df_diwali_sales.isnull().sum()
```

```python
df_diwali_sales.drop('Status', axis=1, inplace=True)
```

```python
df_diwali_sales.drop('unnamed1', axis=1, inplace=True)
```

```python
df_diwali_sales.dropna(inplace=True)
df_diwali_sales
```

```python
df_diwali_sales.isnull().sum()
```

```python
print("Statistical Summary")
print(df_diwali_sales.describe())
```

## 1. Sales distribution as per gender

```python
plt.figure(figsize=(8,8))
df_diwali_sales['Gender'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Sales distribution as per gender ")
new_labels = ['F - Female', 'M - Male']
plt.legend(labels=new_labels, loc="best")
plt.show()
```

## 2. Sales distribution as per Marital Status

```python
plt.figure(figsize=(8,8))
df_diwali_sales['Marital_Status'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Sales distribution as per Marital Status ")
new_labels = ['0 - Married', '1 - Unmarried']
plt.legend(labels=new_labels, loc="best")
plt.ylabel("")
plt.show()
```

## 3. Top 10 Most Purchased Product IDs

```python
# Count number of purchases for every product
top_products = df_diwali_sales.groupby('Product_ID')['User_ID'].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,5))
sns.barplot(x=top_products.index, y=top_products.values, palette = 'rainbow')
plt.xticks(rotation=45)
plt.title("Top 10 Most Purchased Product IDs")
plt.xlabel("Product ID")
plt.ylabel("Number of Purchases (User Count)")
plt.show()
```

## 4. Sales Distribution by Age Group

```python
plt.figure(figsize=(6,6))
df_diwali_sales['Age Group'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Sales Distribution by Age Group") 
plt.ylabel("")
plt.show()
```

## 5. Distribution of Product_ID for Age Group 26-35

```python
age_group_data = df_diwali_sales[df_diwali_sales['Age Group'] == '26-35']

# Group by 'Product_ID' and count the occurrences (distribution)
product_distribution = age_group_data['Product_ID'].value_counts()

# Define a list of 10 distinct colors (the 'tab10' palette)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

# Plot a bar chart for the distribution, using the colorful list
plt.figure(figsize=(10,6))

product_distribution.head(10).plot(kind='bar', color=colors)
plt.title("Distribution of Product_ID for Age Group 26-35")
plt.xlabel("Product_ID")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

 ## 6. State Distribution for Age Group 26–35

```python
age_group_data = df_diwali_sales[df_diwali_sales['Age Group'] == '26-35']

product_distribution = age_group_data['State'].value_counts()

# Define a list of 10 distinct colors (the 'tab10' palette)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

plt.figure(figsize=(10,6))
product_distribution.plot(kind='bar', color=colors)
plt.title("State Distribution for Age Group 26–35")
plt.xlabel("States")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

## 7. Total Sales by Product Category

```python
category_sales = df_diwali_sales.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(y=category_sales.index, x=category_sales.values, palette='rainbow')

plt.ylabel("Product Category")
plt.xlabel("Sum of Amount")
plt.title("Total Sales by Product Category")
plt.show()
```

## 8. Total Sales Amount by States

```python
state_sales = df_diwali_sales.groupby('State')['Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=state_sales.index, y=state_sales.values, palette='rainbow')
plt.xticks(rotation=45)
plt.title("Total Sales Amount by State")
plt.ylabel("Amount")
plt.show()
```

##  9. Orders Distribution by Zone 

```python

df_diwali_sales.groupby('Zone')['Orders'].sum().plot.pie(
    autopct='%1.1f%%',
    figsize=(6,6),
)
plt.title("Orders Distribution by Zone")
plt.legend(loc = "upper right")
plt.ylabel('')
plt.tight_layout()
plt.show()
```

 ## 10. Average Amount per Order by Zone

```python
# Average Amount per Order by Zone
zone_avg = df_diwali_sales.groupby('Zone')['Amount'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,4))
plt.yticks(np.arange(0, 15000, 1000))
sns.barplot(x=zone_avg.index, y=zone_avg.values, palette='rainbow')
plt.title("Average Amount per Order by Zone")
plt.ylabel("Average Amount")
plt.xlabel("Zones")
plt.show()
```

 ## 11. Average Orders per Customer by Occupation

```python
occ_orders = df_diwali_sales.groupby('Occupation')['Orders'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=occ_orders, y='Occupation', x='Orders', palette='rainbow')
plt.xticks(rotation=45)
plt.title("Average Orders per Customer by Occupation")
plt.xlabel("Avg Orders")
plt.ylabel("Occupation")
plt.show()
```

## 12. Trend of Orders Across States

```python
state_orders = df_diwali_sales.groupby('State')['Orders'].sum().sort_values()

plt.figure(figsize=(12,5))
plt.plot(state_orders.index, state_orders.values, marker='*')
plt.xticks(rotation=45)
plt.title("Trend of Orders Across States")
plt.xlabel("States")
plt.ylabel("Total Orders")
plt.grid(True)
plt.show()
```
